```python

object_columns = [column for column in df.columns if df[column].dtype == 'O']


for column in object_columns:
    df[column] = df[column].fillna(df[column].mode()[0])
```

```python
# Select float columns
float_columns = [column for column in df.columns if df[column].dtype == 'float']

# For float values, replace the null values with the mode
for column in float_columns:
    df[column] = df[column].fillna(df[column].mode()[0])
```

# Analysing accidents at different times of the day

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'].str.split('.').str[0], errors='coerce')
df['End_Time'] = pd.to_datetime(df['End_Time'].str.split('.').str[0], errors='coerce')
```

```python
df.info()
```

```python
df['End_Time'].isnull().sum()
```

```python
df['Start_Time'].isnull().sum()
```

```python
df.head()
```

```python
df['Start_Hour'] = df['Start_Time'].dt.hour
```

```python
# Define a function to categorize the time of day
def categorize_time(hour):
    if 4 <= hour < 7:
        return 'Early Morning'
    elif 7 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 20:
        return 'Evening'
    elif 20 <= hour < 24:
        return 'Night'
    else:
        return 'Midnight'
```

```python
df['Time_of_Day'] = df['Start_Hour'].apply(categorize_time)
print(df[['Start_Time', 'Start_Hour', 'Time_of_Day']])
```

```python
accident_counts = df['Time_of_Day'].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x ='Time_of_Day', hue ='Time_of_Day', order=accident_counts.index, palette = 'cividis')
```

# severity of the accident

```python
# 1 indicates the least impact on traffic and 4 indicates a significant impact on traffic.

severity_counts = df['Severity'].value_counts()
labels = severity_counts.index

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(severity_counts, labels=labels, autopct='%1.1f%%', 
        colors=sns.color_palette('coolwarm', len(severity_counts)), pctdistance=0.8)
```

# Accidents based on road conditions

```python
road_condition_columns = ['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 
                          'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']

# Sum the occurrences of accidents for each road condition (True means the condition was present)
road_condition_sums = df[road_condition_columns].sum().sort_values(ascending=False)

# Plot the data
plt.figure(figsize=(20, 10))
sns.barplot(x=road_condition_sums.values, y= road_condition_sums.index, hue=road_condition_sums.index, palette='viridis')
```
